### Dataset #1 – Portland Housing Market Data (2020–2021)

Dataset Description

This dataset contains housing market data for the Portland, Oregon area from 2020 to 2021. The data includes information such as:
	•	City and ZIP code
	•	Number of bedrooms and bathrooms
	•	Days on Zillow (time on market)
	•	Date sold
	•	Zestimate (Zillow’s estimated home value)

The primary variables used for this project are:
	•	zestimate: Estimated housing price (USD)
	•	daysOnZillow: Number of days the property remained on the market
	•	dateSold: Sale date (converted from Unix timestamp)
	•	year/month: Extracted from sale date for time analysis

This dataset allows us to analyze housing market trends before and after the implementation of Measure 110 in Oregon.


```python
import pandas as pd

housing = pd.read_csv("data/00-raw/portland_housing.csv")

housing.head()
```

Cleaning choices:
I reduced the raw Zillow dataset (25731 rows, 348 columns) to a smaller set of relevant variables for my research question: city/zipcode, bedrooms, bathrooms, daysOnZillow, dateSold, and zestimate. Many priceHistory/ and taxHistory/ columns contained substantial missingness and were not needed for our analysis, so I excluded them. I converted dateSold to datetime and coerced invalid entries to missing.

Data Cleaning and Preparation

The dataset originally contained over 300 variables. We reduced the dataset to the most relevant variables for analysis. The following cleaning steps were performed:
	•	Converted Unix timestamp in dateSold to datetime format
	•	Extracted year and month
	•	Converted numerical columns to proper numeric types
	•	Checked missing values
	•	Filtered the data to include only 2020–2021 observations
	•	Saved the cleaned dataset to data/02-processed

```python
housing.shape
```

```python
housing.info()
```

```python
housing.isnull().sum().sort_values(ascending=False).head(10)
```

```python
housing.columns
```

```python
keep_cols = [
    "address/city",
    "address/zipcode",
    "bedrooms",
    "bathrooms",
    "daysOnZillow",
    "dateSold",
    "zestimate",
]

housing_small = housing[keep_cols].copy()

housing_small.head()
```

```python
housing_small.isnull().sum().sort_values(ascending=False)
```

```python
housing_small["dateSold"] = pd.to_datetime(
    housing_small["dateSold"],
    unit="ms",
    errors="coerce"
)
```

```python
housing_small["year"] = housing_small["dateSold"].dt.year
housing_small["month"] = housing_small["dateSold"].dt.month

housing_small[["dateSold", "year", "month"]].head()
```

```python
housing_small.describe()
```

```python
housing_small.groupby("year")["zestimate"].mean()
```

```python
housing_filtered = housing_small[
    (housing_small["year"] >= 2020) &
    (housing_small["year"] <= 2021)
].copy()

housing_filtered.shape
```

```python
import os
os.makedirs("data/02-processed", exist_ok=True)

housing_filtered.to_csv(
    "data/02-processed/housing_clean.csv",
    index=False
)
```

```python
import pandas as pd

overdose = pd.read_csv("data/00-raw/VSRR_Provisional_Drug_Overdose_Death_Counts_20260218.csv")

overdose.head()
```

```python
overdose["State Name"].unique()
```

```python
overdose_or = overdose[overdose["State Name"] == "Oregon"].copy()

overdose_or.head()
```

```python
overdose_or = overdose_or[
    (overdose_or["Year"] >= 2020) &
    (overdose_or["Year"] <= 2021)
].copy()

overdose_or.shape
```

```python
overdose_or["Indicator"].unique()
```

```python
overdose_or[overdose_or["Indicator"].str.contains("Opioid", case=False, na=False)]["Indicator"].unique()
```

```python
overdose_opioid = overdose_or[
    overdose_or["Indicator"] == "Opioids (T40.0-T40.4, T40.6)"
].copy()

overdose_opioid.head()
```

```python
overdose_opioid["date"] = pd.to_datetime(
    overdose_opioid["Year"].astype(str) + "-" + overdose_opioid["Month"],
    errors="coerce"
)
```

```python
overdose_opioid["Data Value"] = pd.to_numeric(
    overdose_opioid["Data Value"],
    errors="coerce"
)
```

```python
overdose_monthly = overdose_opioid.groupby(
    ["Year", "Month"]
)["Data Value"].mean().reset_index()

overdose_monthly.head()
```

```python
import os
os.makedirs("data/02-processed", exist_ok=True)

overdose_monthly.to_csv(
    "data/02-processed/overdose_clean.csv",
    index=False
)
```
